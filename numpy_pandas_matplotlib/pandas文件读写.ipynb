```python
# rank()函数，类似于sql的窗口函数
```

```python
df7=pd.DataFrame({
    '姓名':['胜多负少','挤一挤','路口','ioii'],
    '成绩':[100,44,44,55]
})
```

```python
df7.rank()
```

```python

df7['成绩排名_min']=df7['成绩'].rank(method='min',ascending=False)
df7['成绩排名_max']=df7['成绩'].rank(method='min',ascending=False)
df7['成绩排名_dense']=df7['成绩'].rank(method='dense',ascending=False)
df7['成绩排名_avg']=df7['成绩'].rank(method='average',ascending=False)
df7['成绩排名_moren ']=df7['成绩'].rank(ascending=False)
df7
```
