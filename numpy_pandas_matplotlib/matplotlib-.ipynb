### 1.2添加x，y刻度信息

```python
# 1.创建画布
plt.figure(figsize=(10,5),dpi=100)

#2.主备数据
# plt.plot([1,2,3,4,5,6,7],[14,24,16,3,22,11,26])
x=range(60)
y=[random.uniform(10,20) for i in x]

#3.画图
plt.plot(x,y,color='red')
plt.grid(linestyle='-.',alpha=0.5)


#4.添加刻度信息
x_ticks_label=[f'11点{i}' for i in  x]
y_ticks_label=range(45)

plt.xticks(x[::5],x_ticks_label[::5])
plt.yticks(y_ticks_label[::5])

#添加标题
plt.title('背景某一天一个小时内的气温折线图', fontsize=15)

#添加x轴，y轴标签
plt.xlabel('时间',fontsize=10)
plt.ylabel('时间')

#保存图片
plt.savefig('./my_pic.png')

#4.
plt.show()
```

### 1.3 在一个画布中画多个图

```python
#在一个画布中，绘制两个城市的温度图
#1.准备画布
plt.figure(figsize=(10,5),dpi=100)
#2.准备数据
x=range(60)
y_beijing=[random.uniform(10,20) for i in x ]
y_shanghai=[random.uniform(10,25) for i in x ]

#3添加数据到图中
plt.plot(x,y_beijing,color='blue',label='beijing')
plt.plot(x,y_shanghai,color='red',linestyle='--',label='shanghai')



#4准备标签
x_ticks_label=[f'12点{i} 'for i in x]
y_ticks_label=range(45)

#5添加标签
plt.yticks(y_ticks_label[::5],y_ticks_label[::5])
plt.xticks(x[::5],x_ticks_label[::5])

#6 添加细节
plt.grid(linestyle='-.',alpha=0.5)
plt.title('xxxxxxx',fontsize=20)
plt.xlabel('时间',fontsize=10)
plt.ylabel('温度',fontsize=10)
plt.legend(loc=4)
#6.画图
plt.show()
```

```python
#在一个画布中，绘制两个城市的温度图
#1.准备画布
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5),dpi=100)
#2.准备数据
x=range(60)
y_beijing=[random.uniform(10,20) for i in x ]
y_shanghai=[random.uniform(10,25) for i in x ]

#3添加数据到图中
axes[0].plot(x,y_beijing,color='blue',label='beijing')
axes[1].plot(x,y_shanghai,color='red',linestyle='--',label='shanghai')



#4准备标签
x_ticks_label=[f'12点{i} 'for i in x]
y_ticks_label=range(45)

#5添加标签
#5.1左边
axes[0].set_xticks(x[::5],x_ticks_label[::5])
axes[0].set_yticks(y_ticks_label[::5],y_ticks_label[::5])

axes[1].set_xticks(x[::5],x_ticks_label[::5])
axes[1].set_yticks(y_ticks_label[::5],y_ticks_label[::5])

#6 添加细节

axes[0].set_title('xxxxxxx',fontsize=20)
axes[0].set_xlabel('时间',fontsize=10)
axes[0].set_ylabel('温度',fontsize=10)

axes[1].set_title('xxxxxxx',fontsize=20)
axes[1].set_xlabel('时间',fontsize=10)
axes[1].set_ylabel('温度',fontsize=10)

axes[0].grid(linestyle='-.',alpha=0.5)
axes[1].grid(linestyle='-.',alpha=0.5)
axes[0].legend(loc=4)
axes[1].legend(loc=4)
#6.画图
plt.savefig('./my_pic2.png')
plt.show()
```
