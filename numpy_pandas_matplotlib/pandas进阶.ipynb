# 1.空值处理

```python
# 1. 读取数据
movie_df=pd.read_csv('../data/movie.csv')
movie_df
```

```python
# 2. 查看数据介绍
movie_df.columns
movie_df.info()
movie_df.dropna()
```

```python
#判断数据是否为空
pd.isnull(movie_df)# 判断数值是否为空，空则True
pd.notnull(movie_df)# 非空为True

np.all(pd.notnull(movie_df))
```

```python
# 4.填充缺失值
# movie_df.fillna(23).info()

m1=movie_df['Revenue (Millions)'].fillna(movie_df['Revenue (Millions)'].mean())
m1.info()
```

```python
#5. 循环填充缺失值
for col_name in movie_df.columns:
    # print(np.all(pd.notnull(movie_df[col_name]))==False)
    if np.all(pd.notnull(movie_df[col_name]))==False:
        print(col_name)
        print(movie_df[col_name].mean())
        # movie_df[col_name].fillna(movie_df[col_name].mean(),inplace=True)
        movie_df[col_name].fillna({col_name:movie_df[col_name].mean()})

movie_df.info()
```

```python
for col_name in movie_df.columns:
    # 检查列中是否有缺失值
    if not np.all(pd.notnull(movie_df[col_name])):
        print(col_name)
        mean_value = movie_df[col_name].mean()
        print(mean_value)
        # 避免链式赋值，直接赋值给原DataFrame
        movie_df[col_name] = movie_df[col_name].fillna(mean_value)

movie_df.info()
```

# 2.数据合并

```python
df=pd.read_csv('../data/1960-2019全球GDP数据.csv',encoding='gbk')
df
df1=df[:10]
df2=df[10:20]
```

```python
#2. 通过concat()

pd.concat([df1,df2] ,axis=1)
```

```python
#2.2merge()->只能行合并（水平合并）

df1=pd.DataFrame({
    'Key1':['k0','k0','k1','k2'],
    'Key2':['k0','k1','k0','k1'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})

df2=pd.DataFrame({
    'Key1':['k0','k1','k1','k2'],
    'Key2':['k0','k0','k0','k0'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
#演示merge默认合并方式
pd.merge(df1,df2)
pd.merge(df1,df2,on=['Key1','Key2'],how='inner')
```

```python
# 内连接，指定合并字段
pd.merge(df1,df2,on='Key2',how='inner')
```

# 分组聚合groupby

```python
df=pd.read_csv('../data/uniqlo.csv')
df
```

```python
#agg->aggregate聚合
#格式：df.groupby(['分组字段1','分组字段2',....]).agg({'列名1':'聚合函数名','列名2':'集合函数名',...})
```

```python
# 1. 按照名单列分组
df.groupby(['city'])
df.groupby('city')
```

```python
#2.按照多列分组
df.groupby(['city','channel'])
```

```python
df.groupby(['city','channel']).get_group(('北京','线下'))
```

```python
#4. 分组+聚合
#根据城市和销售渠道分组，计算销售金额
df.groupby(['city','channel']).agg({'revenue':'sum'})
df.groupby(['city','channel']).aggregate({'revenue':'sum'})

df.groupby(['city','channel']).agg({'revenue':'sum','order':'sum'})
```

## 分组过滤

```python
#按照城市分组，查询每组销售金额平均值》200的全部数据
df.groupby('city').revenue.mean()
```

```python
df.groupby('city').filter(lambda x:x.revenue.mean()>200).city
```

# 交叉表

```python
data={
    '性别':['男','女','男','女','男','女','男','女'],
    '购买':['是','否','是','否','是','否','是','否',]
}
df=pd.DataFrame(data)

#创建交叉表
crosstab=pd.crosstab(df['性别'],df['购买'])
print(crosstab)
```

```python
#2.用透视表完成上述的需求
data={
    '性别':['男','女','男','女','男','女','男','女'],
    '购买':['是','否','是','否','是','否','是','否',],
    '金额':[100,150,200,130,160,220,204,894]
}
df=pd.DataFrame(data)
df.pivot_table(index='性别',columns='购买',values='金额',aggfunc="mean")
```
