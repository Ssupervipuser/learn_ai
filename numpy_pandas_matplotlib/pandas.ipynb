## 创建series对象

```python
#创建series对象，采用默认自增索引
s1=pd.Series([1,2,3,4,5])
print(s1)
```

```python
#3.使用字典，元组创建Series对象
#字典
s3=pd.Series({'a':1,'b':2,'c':3})
print(s3)
#元组
s4=pd.Series((11,22,33,44))
print(s4)
```

```python
# 4.使用numpy船舰series对象
s5=pd.Series(np.arange(10))
print(s5)
```

# DataFrame对象

## 3.1创建DataFrame对象

```python
#1.通过字典+列表方式实现
#1.1准备数据集，每个键值对=1  列  数据
info={
    'name':['aaa','bbb','ccc'],
    'gender':['male','female','male'],
    'age':[33,44,55]
}
#1.2 封装成DataFrame对象
df1=pd.DataFrame(data=info)
print(df1,type(df1))
```

```python
#2. 通过列表+元组方式实现
#1.每个元组=一  行  数据
info=[
    ('aaa','female',33),
    ('bbb','male',66),
    ('ccc','female',66),
]
#2.封装成DataFrame对象
df2=pd.DataFrame(data=info,columns=['name','gender','age'])
print(df2,type(df2))
```

```python
#3.通过numpy的ndarray->pandas DataFrame的方式实现
#1.创建numpy的ndarray对象
arr1=np.arange(12).reshape(3,4)
arr1
```

```python
#2.封装DataFrame对象
df3=pd.DataFrame(data=arr1,columns=['a','b','c','e'])
df3
```

## 逻辑运算符

```python
df
```

```python
# 筛选出open》23的数据
df[df.open>23]

# 帅选出open》23 且《24的数据
df[(df.open>23) & (df.open<24)]
```

```python
# 通过query（）
df.query('open>23& open<24')
```

```python
# 固定值的筛选 isin
# 查询open价格为23.53  23.85
df.query('open in [23.53 , 23.85]')
df[df["open"].isin([23.53, 23.85])]
```

## 统计函数·

```python
# count(),mean(),median(),min(),std(),var()
df.count()
df.mean()
df.median()
df.min()
df.std()
df.var()
```

```python
#apply() 自定义函数
#df.apply(自定义函数，axis=0）


#同时获取到多列的最大值和最小值的差值, open close
#1. 分解
def my_fun(col):
    return col.max()-col.min()
#2.通过apply（）调用函数，作用到指定的列
df[['open','close']].apply(my_fun,axis=0)
df[['open','close']].apply(my_fun,axis=1)



#lambda表达式
df[['open','close']].apply(lambda x: x.max()-x.min())
```
