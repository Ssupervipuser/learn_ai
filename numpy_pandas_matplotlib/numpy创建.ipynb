## 统计函数

```python
#1. 创建ndarray对象
arr=np.arange(12).reshape(3,4)
print(arr)

```

```python
#cumsum累加和 ，cumprod累乘积
a_cumsum=np.cumsum(arr)
print(a_cumsum)
```

```python
print(np.cumprod(arr))
```

```python
print(np.sum(arr))
```

```python
print(np.mean(arr))
```

```python
print(np.max(arr))
```

```python
print(np.min(arr))
```

```python
print(np.std(arr))
```

## 矩阵运算

```python
#1.定义数组
arr1=np.array([10,20,30,40])
arr2=np.arange(4)
print(f'arr2:{arr1}')
print(f'arr2:{arr2}')
```

```python
#减法运算
print(arr1-arr2)
```

```python
#加法运算
print(arr1+arr2)
```

### 矩阵乘法

```python
#行列数一致
print(arr1@arr2)
print(np.dot(arr1,arr2))
print(arr1.dot(arr2))
# print(arr1*arr2)
```

```python
arr3=np.array([
    [1,2,3],[4,5,6]
])
arr4=np.array([[8,9],[6,5],[3,2]])
print(arr3)
print(arr4)

#2.行列数不一致
print(np.dot(arr3,arr4))
```
