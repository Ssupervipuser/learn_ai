```python

import pandas as pd
```

```python
data = {
    "order_id": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    "user_name": ["张三", "李四", "王五", "赵六", "钱七", "张三", "李四", "孙八"],
    "product": ["手机", "电脑", "耳机", "手机", "平板", "电脑", "耳机", "平板"],
    "price": [3999, 5999, 199, 3899, 2499, 5899, 189, 2399],
    "quantity": [1, 1, 2, 1, 1, 1, 3, 1],
    "pay_type": ["支付宝", "微信", "支付宝", "微信", "支付宝", "微信", "支付宝", "微信"]
}

df = pd.DataFrame(data)
df
```

```python
# 1. 查看数据集前5行
#    知识点：head()
df.head()
# 1. 查看后3行
#    知识点：tail()
df.tail(3)
# 1. 获取 user_name 列（两种写法）
#    知识点：单列获取
df['user_name']
df.user_name
# 1. 获取 product、price 两列
#    知识点：多列获取
df[['product', 'price']]
# 1. 获取第2行到第4行
#    知识点：行切片
df.iloc[2:4, :]
# 1. 筛选 product == "手机"
#    知识点：布尔索引
df.query("product=='手机'")
df[df['product'] == '手机']
# 1. 筛选 price > 3000
#    知识点：数值筛选
df.query('price>3000')
# 1. 筛选 user_name == "张三"
#    知识点：精确匹配
df.query('user_name=="张三"')
# 1. 使用 query 筛选支付方式为微信
#    知识点：query
df.query('pay_type=="微信"')
# 1. 筛选产品为电脑且价格>5000
#    知识点：多条件与
df.query('product=="电脑"&price>5000')
df[(df['product']=='电脑')&(df['price']>5000)]
type(df['product']=='电脑')
type(df.product=='电脑')
```

```python
df.info()
```

```python
# 1. 新增 discount = 0.9
#    知识点：直接赋值新增列
df['discount']=0.9
df
# 1. 新增 total = price * quantity
#    知识点：列运算新增列
df['total']=df['price']*df['quantity']
df
# 1. assign 新增 tax = 0.06
#    知识点：assign
df=df.assign(tax=0.06)
df
# 1. assign 同时新增 mark=0、level="普通"
#    知识点：assign多列
df=df.assign(mark=0,level='普通')
df
# 1. 新增 is_expensive：≥3000为1，否则0
#    知识点：apply条件列
# df.apply(lambda is_expensive>=3000:1,0)
def fun(price):
    if price>=3000:
        return 1
    else:
        return 0

df['is_expensive'] = df['price'].apply(fun)
df
df['is_expensive']=df['price'].apply(lambda x:1 if x>=3000 else 0)
df
# 1. 新增 buy_num = [1,1,2,1,1,1,3,1]
#    知识点：列表赋值列
df['buy_num']=[1,1,2,1,1,1,3,1]
# 1. 新增 pay_month = "2026-04"
#    知识点：固定字符串列
df['pay_month']='2026-04'
# 1. assign 新增 final_price = total * discount
#    知识点：assign+运算
df=df.assign(final_price=df.total*df.discount)
# 1. 新增 product_type：手机/电脑为数码，其他为配件
#    知识点：自定义函数apply
df['product_type']=df['product'].apply(lambda x:'数码' if x=='手机'or x=='电脑' else '其他配件')
# 1. 复制 price 为 price_copy
#    知识点：复制列
df['price_copy']=df.price.copy()
```

```python
# 1. 修改第0行 price 为 3999
#    知识点：loc单值修改
df.loc[0,'price']=3999
# 1. 把所有耳机的 price 改为 199
#    知识点：条件批量修改

# df['price']=df['product'].apply(lambda x:199 if x=='耳机')
# def fun(product):不对
#     if product=='耳机':
#         return 199
# df=df.assign(price=fun(df.product))
df.loc[df['product'] == '耳机', 'price'] = 199

# 1. replace 支付宝→AliPay
#    知识点：replace替换
# df.method({col: value}, inplace=True)
df.replace({'支付宝':'AliPay'},inplace=True)
# 1. replace 微信→WeChat，支付宝→AliPay
#    知识点：多值替换
df.replace({'支付宝':'AliPay1','微信':'WeChat'},inplace=True)
# 1. quantity 全部 +1
#    知识点：整列数值修改
df['quantity']=df['quantity']+1
# 1. 列名 pay_type → payment
#    知识点：rename
df.rename(columns={'pay_type':'payment'},inplace=True)
# 1. 订单1003的用户名改为小王
#    知识点：条件定位修改
df.loc[df['order_id']==1003,'user_name']='小王'
# 1. assign 覆盖 total = pricequantity0.95
#    知识点：assign覆盖列
df=df.assign(total=df['price']*df['quantity']*0.95)
# 1. quantity>2 标记 is_bulk=1
#    知识点：条件标记
df['is_bulk']=df['quantity'].apply(lambda x:1 if x>2 else 0)
# df['is]=df.loc[:,df['quantity'>2]]
# 1. 张三→张三丰
#    知识点：字符串replace
df.replace({'张三':'张三丰'},inplace=True)
df
```
