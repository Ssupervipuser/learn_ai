```python
plt.figure(figsize=(20,8),dpi=80)
x=range(60)
#准备两组y轴数据
y_bj=[random.uniform(5,18) for i in x]
y_sh=[random.uniform(8,18) for i in x]



x_ticks_label=[f'11点{i}分' for i in x]
y_ticks=range(20)

plt.xticks(x[::5],labels=x_ticks_label[::5])
plt.yticks(y_ticks[::2])

plt.title('sfsfsfs',fontsize=20)
plt.xlabel('时间',size=200)
plt.ylabel('温度',size=12)
#网格
plt.grid(linestyle='--',alpha=0.5)
#图例
plt.legend(loc=1)
#画图
plt.plot(x,y_bj,color='pink',label='beijing')
plt.plot(x,y_sh,color='blue',label='shanghai')

plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

x=range(60)

y_bj=[random.uniform(5,18) for i in x]
y_sh=[random.uniform(8,18) for i in x]


axes[0].plot(x,y_bj,label='bj',linestyle=':',color='m')
axes[1].plot(x,y_sh,label='sh',linestyle='--')

x_ticks_label=[f'11点{i}分' for i in x]
y_ticks=range(20)

axes[0].set_xticks(x[::5],labels=x_ticks_label[::5])
axes[0].set_yticks(y_ticks[::2])

axes[1].set_xticks(x[::5],labels=x_ticks_label[::5])
axes[1].set_yticks(y_ticks[::2])

axes[0].set_title('bjbjbjbjbjbj',fontsize=20)
axes[0].set_xlabel('时间',fontsize=12)
axes[0].set_ylabel('温度',fontsize=12)


axes[1].set_title('shshsshshsh',fontsize=20)
axes[1].set_xlabel('时间',fontsize=12)
axes[1].set_ylabel('温度',fontsize=12)

axes[0].grid(linestyle='--',alpha=0.5)
axes[1].grid(linestyle='-.',alpha=0.5)


axes[0].legend(loc='best')
axes[1].legend(loc='best')

plt.show()
```
