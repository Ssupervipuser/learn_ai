```python
df=pd.read_csv('../data/1960-2019全球GDP数据.csv', encoding='gbk')
# print(type(df))
df
```

```python
china_df=df[df.country=='中国']
usa_df=df[df.country=='美国']
jp_df=df[df.country=='日本']
print(type(china_df))
```

```python
china_df.set_index('year',inplace=True)
usa_df.set_index('year',inplace=True)
jp_df.set_index('year',inplace=True)
china_df
```

```python
china_df.GDP.plot(legend='中国')
usa_df.GDP.plot(legend='美国')
jp_df.GDP.plot(legend='日本')
```

```python
china_df.rename(columns={'GDP':'china'},inplace=True)
usa_df.rename(columns={'GDP':'usa'},inplace=True)
jp_df.rename(columns={'GDP':'jp'},inplace=True)
```

```python
china_df.china.plot(legend='中国')
usa_df.usa.plot(legend='美国')
jp_df.jp.plot(legend='日本')
```
